```python
import tensorflow 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Input
import numpy as np
```

```python
model = Sequential(
    [
        Input(shape=(2,)),
        Dense(2, activation="linear"),
        Dense(1, activation="linear")
    ]
)
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_weights = [
    np.array([[0.1, 0.1],
              [0.1, 0.1]], dtype=np.float32),   # W1 — shape (2,2)
    np.array([0., 0.], dtype=np.float32),        # b1 — shape (2,)
    np.array([[0.1],
              [0.1]], dtype=np.float32),          # W2 — shape (2,1)
    np.array([0.], dtype=np.float32)             # b2 — shape (1,)
]

model.set_weights(new_weights)
```
